```python
logsLoc = "logs/"
logs = os.listdir(logsLoc)
```

```python
for log in logs:
    if log.split('.')[-1]=='log':
        model_stats = pd.read_csv(logsLoc+log,engine='python')
        plt.figure()
        plt.plot(model_stats.val_categorical_accuracy)
        plt.title(log)
        plt.annotate('max val acc: {}'.format(max(model_stats.val_categorical_accuracy)), xy=(0.5,0.5),xytext=(.5, .1), textcoords='axes fraction')
```
